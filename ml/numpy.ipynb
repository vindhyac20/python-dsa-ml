```python
#numpy 
#1.more memory used by list than numpy array
import numpy as np
size=10
numpy_array=np.arange(size)

print(numpy_array.nbytes)
```

```python
#list
import sys
size=10
list_py=list(range(size))
print(sys.getsizeof(list_py[0])*len(list_py))
```

```python
#less time to access compared to list
import time
l1=[1,2,3,4]
l2=[3,5,6,8]
start_time=time.time()
l3=[x+y for x,y in zip(l1,l2)]
end_time=time.time()
print(end_time-start_time)
print(l3)
print(l1+l2)

```

```python
import time
import numpy as np
l1=np.array([1,2,3,4])
l2=np.array([3,5,6,8])
start_time=time.time()
l3=l1+l2
end_time=time.time()
print(end_time-start_time)
print(l3)
```
