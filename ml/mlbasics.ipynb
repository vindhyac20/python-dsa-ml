```python
#iqr
import numpy as np
math_data=np.array(data['math score'])
q1=np.percentile(math_data,25)
print(q1)
q3=np.percentile(math_data,75)
print(q3)
iqr=q3-q1
print(iqr)

```

```python
#lower 
lower=q1-(1.5*iqr)
print(lower)
upper=q3+(1.5*iqr)
print(upper)
```

```python
lower_outliers=np.where(data['math score']<lower)[0]
print(lower_outliers)
print(len(lower_outliers))
upper_outliers=np.where(data['math score']>upper)[0]
print(upper_outliers)
print(len(upper_outliers))
```
