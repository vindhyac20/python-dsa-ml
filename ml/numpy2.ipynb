```python
import numpy as np

X = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

b = X[0]        # b is a VIEW of row index 0 → [10, 20, 30]
```

```python
c = X[1].copy()   # c is an independent COPY of row index 1
c[0] = -5
print(X)          # X is untouched — still has 40 in that spot
print(c)          # [-5, 50, 60]
```

```python
scores = np.array([[80, 92, 70], [95, 60, 88], [72, 85, 91]])
students=scores.mean(axis=1)
print(students)
sub=scores.mean(axis=0)
print(sub)
top=scores.argmax(axis=0)
print(top)
```
