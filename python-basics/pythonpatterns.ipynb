```python
# while loop    # IF WE ENTER 0 PRINCIPLE OR ANYTHING THE CODE WILL BE WRONG, SO WE USE (while True ) which means the condition is true always
principle=0
rate=0
time=0

while  principle <= 0:
    principle=int(input("enter the principle"))
    if principle<=0:
        print("enter valid number")
print(f"principle is {principle}")
while  rate <= 0:
    rate=int(input("enter the rate"))
    if rate<=0:
        print("enter valid number")
print(f"rate is {rate}")
while  time <= 0:
    time=int(input("enter the time"))
    if time<=0:
        print("enter valid number")
print(f"rate is {time}")
amount=principle*((1+rate/100)**time)
print(amount)


```
