```python
#inheritance
class Animal:
    def __init__(self,name):
        self.name=name
    def eat(self):
        print(f"{self.name} is eating")
    def sleep(self):
        print(f"{self.name} is sleeping")
#inheritance
class Dog(Animal):
    def speak(self):
        print("bark")
class Cat(Animal):
    def speak(self):
        print("meow")
dog=Dog("stephen")
cat=Cat("garett")
dog.eat() #no need to use print 
dog.speak()
cat.eat()
    
    
            
        
        
```

```python
class Animal:

    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

class Prey(Animal):
    def flee(self):
        print(f"{self.name} is fleeing")

class Predator(Animal):
    def hunt(self):
        print(f"{self.name} is hunting")

class Rabbit(Prey):
    pass

class Hawk(Predator):
    pass

class Fish(Prey, Predator):
    pass

rabbit = Rabbit("Bugs")
hawk = Hawk("Tony")
fish = Fish("Nemo")
fish.hunt()
```
