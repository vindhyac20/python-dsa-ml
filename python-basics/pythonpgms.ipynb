```python
#timer
import time
timer=int(input("enter the seconds"))

for x in range(timer,0,-1):
    seconds=x%60
    minutes=int(x/60)%60
    hours=int(x/3600)
    print(f"{ hours:02}:{ minutes:02}:{seconds:02}")
    time.sleep(1)
print("time up")
    
    


```

```python
#shopping cart
cart_item={"pizza":1.22,"hotdog":3.44,"lemonade":2.11,"milkshake":3.44}
total=0
cart=[]
print("------------menu-----------")
for key,values in cart_item.items():
    print(f"{key:10}:{values}")
print("----------------------------")
is_running=True
while is_running:
    order=input("enter the items to order(q to quit):")
    if order.lower()=="q":
        print("thank You")
        is_running=False
    elif cart_item.get(order) is not None:  # if a key is not in dictionary it returns None
        cart.append(order)
for order in cart:
    total+=cart_item.get(order)
    print(f"total is:{total}")

    
        

    
        
```

```python
#number guessing game
import random
choose=random.randint(1,100)
trials=0
is_running=True
while True:
    guess=int(input("enter the number between 1-100"))
    if guess==choose :
        trials+=1
        print("correct answer")
        break
    elif guess<choose:
        trials+=1
        print("too low")
    elif guess>choose:
        trials+=1
        print("too high")
print(int(trials))

    
              

```

```python
#*args -pass multiple non keyword arguments = tuple
def person_name(*args):
    for arg in args:
        print(arg,end=' ')
person_name ("mr","chiku","kohli")
```

```python
#**kwargs = dict
def shipping_address(**kwargs):
    for value in kwargs.values():
        print(value)

shipping_address(house_no=2,street="fake street",city="london")
```

```python
#match case (switch in other language)    * case replace with the parameter name*  | is or & is and
def is_weekend(week_name):  
    match week_name:
        case "sunday"|"saturday":
            print("its weekend")
        case "mon"|"tue"|"wed"|"thur"|"fri":
            print("its weekday")
is_weekend("mon")
        
```
