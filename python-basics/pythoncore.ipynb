```python
#encryption and decryption pgm
import string
import random
#print(help('string'))
chars=" "+string.punctuation+string.digits+string.ascii_letters
chars=list(chars)
print(chars)
keys=chars.copy()
#(f"keys={keys}")
random.shuffle(keys)
print(f"shuffled :{keys}")
#encrypt
plain_text = input("enter the sentence")
cipher_text=""
for letter in plain_text:    
    index=chars.index(letter)   #
    cipher_text+=keys[index]
print(cipher_text)  

#decrypt
cipher_text=input("enter cipher_text")
plain_text=''
for letter in cipher_text:
    index=keys.index(letter)
    plain_text+=chars[index]
print(plain_text)



    

    
    
    
    
    
    
```
