```python
nums = [2, 7, 11, 15]
target = 9
left = 0
right = len(nums) - 1

while left < right:
    current_sum = nums[left] + nums[right]
    
    if current_sum == target:
        print([left, right])
        break
    elif current_sum < target:
        left += 1       # too small, need bigger → move left right
    else:
        right -= 1      # too big, need smaller → move right left
```
