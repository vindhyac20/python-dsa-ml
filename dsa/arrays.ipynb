```python
#second largest element in the array  o(n)
def second_largest(nums):
    largest = float('-inf')
    second = float('-inf')
    
    for n in nums:
        if n > largest:
            second = largest    # old largest drops down
            largest = n         # new largest takes over
        elif n > second and n != largest:
            second = n          # new second largest
    
    return second

print(second_largest([3, 1, 5, 2, 4]))   # 4
print(second_largest([10, 10, 9]))         # 9


```
