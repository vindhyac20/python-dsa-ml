```python
from collections import defaultdict
l = defaultdict(list)
group = [1,2,3,4,5,6,7,8]
for number in group:
    if number % 2 == 0:
        l["even"].append(number)
    else:
        l["odd"].append(number)
print(l["even"])
print(l["odd"])
```

```python
nums = [1,12,-5,-6,50,3]
k = 4

# step 1: first window
current = sum(nums[:k])
best = current

# step 2: slide
for i in range(k, len(nums)):
    current = current - nums[i-k] +nums[i]
    best = max(best,current)

# step 3: return average
print(best)
```
