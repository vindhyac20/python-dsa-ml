```python
#reverse the number
num=6789283
reverse=0
while num>0:
    last_digit=num%10
    reverse=(reverse*10)+last_digit
    num=num//10
print(reverse)
```

```python
#palindrome
num=121
n=num   #a duplicate of num should be compared beacuse num is altered
reverse=0
while num>0:
    last_digit=num%10
    reverse=(reverse*10)+last_digit
    num=int(num/10)  # or use num=num//10  
if n == reverse:
    print("it is palindrome")
else:
    print("not")
```

```python
#armstrong number  -taking sum of power of total number of digits to be equal to the number
num=1634
n=num   #a duplicate of num should be compared beacuse num is altered
sum=0
cntdigits=4
while num>0:
    last_digit=num%10
    sum+=pow(last_digit,cntdigits)
    
    num=num//10  # or use num=num//10* 
if n==sum:
    print("it is armstrong")
else:
    print("its not")
    
```
